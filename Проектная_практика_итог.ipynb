```python
# 2. Импортируем библиотеки
from inference_sdk import InferenceHTTPClient
import json
```

```python
# 3. Инициализация клиента в Roboflow
# Создаём экземпляр клиента для взаимодействия с API
client = InferenceHTTPClient(
    # Указываем URL сервера Roboflow для serverless-инфраструктуры
    api_url="https://serverless.roboflow.com",
    # Передаём API-ключ для аутентификации
    api_key="_____" # <- Вставьте сюда Ваш API-ключ
)

# 4. Запускаем рабочий процесс для изображения
# Вызываем метод для запуска workflow (рабочего процесса) на сервере
result = client.run_workflow(
    # Указываем имя рабочего пространства в Roboflow
    workspace_name="elizavetank",
    # Указываем ID workflow (модели) для обнаружения медведей
    workflow_id="detect-bears",
    # Передаем словарь с данными об изображении
    images={
        # Путь к файлу изображения
        "image": "_____.jpg" # <- Вставьте сюда ссылку на Ваше изображение
    },
    # Включаем кэширование для ускорения повторных запросов с тем же изображением
    use_cache=True
)

# 5. Выводим информацию о медведях на изображении
# Проверяем, что результат существует, является списком и не пуст
if result and isinstance(result, list) and len(result) > 0:
    # Берём первый элемент списка результатов
    item = result[0]
    # Извлекаем значение поля 'count' (количество объектов), если его нет — возвращаем 0
    count = item.get('count', 0)
    # Инициализируем переменную для класса объекта со значением по умолчанию
    class_name = "неизвестен"
    # Проверяем наличие вложенных ключей
    if 'predictions' in item and 'predictions' in item['predictions']:
        # Получаем список предсказаний (детектированных объектов)
        predictions_list = item['predictions']['predictions']
        # Проверяем, что список не пуст
        if predictions_list and len(predictions_list) > 0:
            # Извлекаем класс первого объекта, если нет — возвращаем "неизвестен"
            class_name = predictions_list[0].get('class', "неизвестен")
    # Выводим количество обнаруженных медведей
    print(f"На изображении обнаружено медведей: {count}")
else:
    # Если результат пустой или некорректный — сообщаем, что медведь не найден
    print("Медведь не обнаружен")
```

```python
# 6. Вывод всего результата (без поля 'annotated_image', потому что оно выдаёт длинное нечитаемое значение)
# Создаём пустой список, куда будем складывать отфильтрованные элементы результата
filtered_result = []
# Проходим по каждому элементу в исходном результате
for item in result:
    # Для текущего элемента создаём новый словарь, в который копируем все пары «ключ‑значение»,
    # кроме той, где ключ равен 'annotated_image'
    filtered_item = {k: v for k, v in item.items() if k != 'annotated_image'}
        # Добавляем отфильтрованный словарь в итоговый список
    filtered_result.append(filtered_item)
# Выводим отфильтрованный результат в формате JSON с отступами (indent=2) для читаемости
print(json.dumps(filtered_result, indent=2))
```
